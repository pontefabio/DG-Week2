```python
# Define age groups 
age_bins = [15, 30, 40, 55, float('inf')]  # Age groups: 15-30, 30-40, 40-55, 55+
age_labels = ['15-30', '30-40', '40-55', '55+']

# Segment customers into age groups
CompaniesDataCity['Age Group'] = pd.cut(CompaniesDataCity['Age'], bins=age_bins, labels=age_labels, right=False)

# Group by 'Company', 'Age Group', and resample by month, calculating profit
monthly_profit_age_grouped = CompaniesDataCity.groupby(['Company', 'Age Group', pd.Grouper(key='date', freq='M')])['Price Charged', 'Cost of Trip'].sum().reset_index()
monthly_profit_age_grouped['Profit'] = monthly_profit_age_grouped['Price Charged'] - monthly_profit_age_grouped['Cost of Trip']

# Define a color palette for the plot
palette = sns.color_palette("husl", len(monthly_profit_age_grouped['Age Group'].unique()))

# Create separate plots for each age group
for company, data in monthly_profit_age_grouped.groupby('Company'):
    # Create a new figure and axis
    fig, ax1 = plt.subplots()
    
    # Plot profit for each age group on the left y-axis
    for i, (age_group, subdata) in enumerate(data.groupby('Age Group')):
        ax1.plot(subdata['date'], subdata['Profit'], linestyle='dashed', label=f"{age_group} Profit", color=palette[i])

    # Set y-axis label for profit
    ax1.set_ylabel('Profit')

    # Add labels and legend
    ax1.set_xlabel('Date')
    ax1.set_title(f'Profit over Time by Age Group for {company}')
    ax1.legend(loc='upper left')

    # Tilt x-axis labels by 45 degrees
    plt.xticks(rotation=45)

    # Show plot
    plt.show()
```

```python

# Group by 'Company' and 'City', and calculate total profit and total kilometers traveled
profit_city_grouped = CompaniesDataCity.groupby(['Company', 'City']).agg({'Price Charged': 'sum', 'Cost of Trip': 'sum', 'Population': 'first', 'KM Travelled': 'sum', 'Customer ID': 'count'}).reset_index()
profit_city_grouped['Profit'] = profit_city_grouped['Price Charged'] - profit_city_grouped['Cost of Trip']
profit_city_grouped['Avg KM Travelled'] = profit_city_grouped['KM Travelled'] / profit_city_grouped['Customer ID']

# Sort the data by profit in descending order
profit_city_grouped = profit_city_grouped.sort_values(by='Profit', ascending=False)

# Define a color palette for the plot
palette = sns.color_palette("husl", len(profit_city_grouped['City'].unique()))

# Create separate plots for each company
for company, data in profit_city_grouped.groupby('Company'):
    # Create a new figure and axis
    plt.figure(figsize=(10, 6))
    
    # Plot profit for each city
    ax = sns.barplot(x='City', y='Profit', data=data, palette=palette, order=data['City'])
    
    # Set labels and title
    plt.xlabel('City')
    plt.ylabel('Profit')
    plt.title(f'Profit by City for {company}')
    
    # Rotate x-axis labels by 45 degrees for better readability
    plt.xticks(rotation=45)
    
    # Add population label to each bar
    for i, bar in enumerate(ax.patches):
        population = data.iloc[i]['Population']
        
        # Add population label above the bar
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05 * ax.get_ylim()[1], f'Population: {population}', ha='center', va='bottom')
    
    # Show plot
    plt.show()
```
